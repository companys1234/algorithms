```python
x,y = map(int,input().split())
m = 1000000007
n = int(input())
if n ==1:
  print(x%m)
if n == 2:
  print(y%m)
seq = [x % m, y % m]
for i in range(2, 6):
    seq.append((seq[i-1] - seq[i-2]) % m)
ans = seq[(n - 1) % 6]
print(ans)
```

```python
x,y = map(int,input().split())
m = 1000000007
n = int(input())
if n ==1:
  print(x%m)
if n == 2:
  print(y%m)
if n > 2:
  c = y
  g = x
  for i in range(n-2):
    z = c - g
    g = c
    c = z
  print(c%m)
```

```python
x, y = map(int, input().split())
n = int(input())

# Последовательность с периодом 6
cycle = [x, y, y - x, -x, -y, x - y]

# Приводим n к индексу в цикле (нумерация с 0)
result = cycle[(n - 1) % 6]

# Выводим результат по модулю 10^9 + 7
print(result % 1000000007)
```
