```python
def print_all_substrings(s):
    n = len(s)
    for i in range(n):  # Начальный индекс подстроки
        for j in range(i, n):  # Конечный индекс подстроки
            substring = s[i:j+1] # Извлекаем подстроку от i до j включительно
            print(substring)

# Пример использования
text = "abacaba"
print_all_substrings(text)
```

```python
n, k = map(int, input().split())
s = input()
allowed = set(input().split())

result = 0
current_length = 0

for char in s:
    if char in allowed:
        current_length += 1
    else:
        # Добавляем количество подстрок для текущего отрезка
        result += current_length * (current_length + 1) // 2
        current_length = 0

# Добавляем последний отрезок (если строка заканчивается разрешенными буквами)
result += current_length * (current_length + 1) // 2

print(result)
```
