```python
t = int(input())
for _ in range(t):
    n, s = map(int, input().split())
    a = list(map(int, input().split()))

    total_ones = sum(a)

    if s > total_ones:
        print(-1)
        continue

    if s == total_ones:
        print(0)
        continue

    # Ищем максимальную длину подмассива с суммой s
    left = 0
    current_sum = 0
    max_len = 0

    for right in range(n):
        current_sum += a[right]

        while current_sum > s:
            current_sum -= a[left]
            left += 1

        if current_sum == s:
            max_len = max(max_len, right - left + 1)

    if max_len == 0:
        print(-1)
    else:
        print(n - max_len)
```
