```python
t = int(input())
for _ in range(t):
  a,k = map(int,input().split())
  an = a
  for i in range(k-1):
    digits = [int(digit) for digit in str(an)]
    e = an + min(digits) * max(digits)
    an = e
  print(an)
```

```python
t = int(input())
for _ in range(t):
  a,k = map(int,input().split())
  an = a
  for i in range(k-1):
    digits = [int(digit) for digit in str(an)]
    e = an + min(digits) * max(digits)
    if min(digits) == 0:
        an = e
        break
    an = e
  print(an)
```
